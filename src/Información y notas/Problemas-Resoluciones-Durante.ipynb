* El original hacía df_tracks.dropna(inplace=True) eliminando cualquier fila con un solo NaN aunque fuera en columnas no críticas. Solución: dropna(subset=['mbid']) para df_tracks y dropna(subset=['name', 'artist']) para df_clean. Se explica el impacto en comentario.

#6 — Variables indefinidas (df en lugar de df_clean, plays_totales_artista inexistente)

#8 — kruskal() con groupby('genre')
